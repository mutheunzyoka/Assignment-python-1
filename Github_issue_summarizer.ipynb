```python
from transformers import pipeline
import gradio as gr
import torch
```

```python
summarizer =  pipeline("summarization" , model="facebook/bart-large-cnn")
```

```python
def summarizer_issue(text):
  if len(text.strip()) <30:
    return "Please enter a longer text"
  result = summarizer(text,max_length=60,min_length = 30 ,do_sample=False)
  return result[0]['summary_text']
```

```python
gr.Interface(
    fn=summarizer_issue,
    inputs= gr.Textbox(lines=10,placeholder="Paste a GitHub issue here..."),
    outputs="textbox",
    title="GitHub Issue Summarizer",
    description="Summarize long GitHub issues using a BART transformer model."
    ).launch()

```
